```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
df = df.iloc[:,2:]
# iloc[:,2:] means take all rows and take columns from 3rd column onwards
```

```python
df.sample(5)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis=1),
                                                    df['Purchased'],
                                                    test_size = 0.3,
                                                    random_state=0)
# df.drop('Purchased',axis=1) - we don't want  purchased column in  x that's why and axis=1 means drop column not row
# random_state = 0 - give the same shuffling every time(we can put any integer in place of 0)
```

```python
X_train.shape, X_test.shape
```

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# fit the scaler to train set, it will learn parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
# means are 37.864.. and 69807.14... resp.
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
# we got numpy arrays from scaler instead of dataframes and hence we are converting them here
```

```python
np.round(X_train.describe(), 1)  # rounding all the values to 1 place
```

```python
np.round(X_train_scaled.describe(),1)
```

# Effect of Scaling

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
# Comparison of Distributions
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Age distribution before scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

ax2.set_title('Age distribution after scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
# Comparison of Distributions
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Salary distribution before scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('Salary distribution after scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

# Why scaling is important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("Actual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```
